```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv('mxmh_survey_results.csv')

dataset = data.fillna(0)

X1 = dataset[['Anxiety']].values
y = dataset['Depression'].values


linreg =LinearRegression()
linreg.fit(X1, y)






import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
plt.plot(X1, y, "b.")
plt.plot(X1, linreg.predict(X1), color='red', label='Linear Regression')
plt.xlabel("$Axiety$")
plt.ylabel("$Depression$", rotation=0)
plt.axis([0, 10, 0, 10])
plt.grid()
plt.show

```
